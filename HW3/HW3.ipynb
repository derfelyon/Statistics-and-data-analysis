```python
def make_p_dis(r,n):
    
    data = np.random.multivariate_normal([0, 0], [[1, r], [r, 1]], size=n)
    X, Y = data[:,0], data[:,1]
    return (X, Y)
```

```python
A,B = make_p_dis(0.2,49)
print('Data with n-1 point pearsopn of', stats.pearsonr(A, B)[0])
A = np.insert(A,-1,50)
B = np.insert(B,-1,-50)
print('Data with all of the 50 data poiny, pearson of', stats.pearsonr(A, B)[0])
sb.jointplot(A,B)
```

#### 1.B
Data with τ(x,y) > ρ(x,y) + 0.45

```python
def biger():
    
    A = np.arange(1,51,1)
    for i in range(1,51,1):
        B1 = np.arange(i,51,1) 
        B2 = np.arange(1,i,1)
        B = np.concatenate((B1, B2), axis=None)
        if ((kendalltau(A, B)[0] )> (stats.spearmanr(A, B)[0] + 0.45)):
            return(A,B)
    return(None,None)
    

A,B = biger()
print('kendal is = ', kendalltau(A, B)[0], 'spearman is =' , stats.spearmanr(A, B)[0] )
sb.jointplot(A,B)
```

```python
from scipy.stats import multivariate_normal as mn
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def plot_2d_Gaussian_pdf(means, cov):
    n = 100
    x1 = np.linspace(means[0] - 3 * np.sqrt(cov[0][0]), means[0] + 3 * np.sqrt(cov[0][0]), n)
    x2 = np.linspace(means[1] - 3 * np.sqrt(cov[1][1]), means[1] + 3 * np.sqrt(cov[1][1]), n)
    x1_v, x2_v = np.meshgrid(x1, x2)
    Xgrid = np.vstack([x1_v.ravel(), x2_v.ravel()]).T
    Y = mn.pdf(Xgrid, means, cov)
    fig, ax = plt.subplots()
    ax.pcolorfast(x1, x2, Y.reshape(x1_v.shape), alpha=0.5, cmap='Blues')
    ax.contour(x1_v, x2_v, Y.reshape(x1_v.shape),
               levels=[0.05, 0.1, 0.15, 0.2], alpha=0.3, colors='b')
    ax.axis('equal')
    ax.grid(alpha=0.2)
    plt.show()
    
means = [54.366337, 149.646865]
cov = [[82.48455839, -82.90331782], [-82.90331782, 524.6464057]]
plot_2d_Gaussian_pdf(means, cov)
```

*A distribution is said to have a heavy right tail if its tail probabilities vanish slower than any exponential*
$$ \forall t>0, \lim_{x\to\infty} e^{tx}P(X>x)=\infty $$

#### 3.A

$${\displaystyle p(x)={\frac {1}{\sigma {\sqrt {2\pi }}}}e^{-{\frac {1}{2}}\left({\frac {x-\mu }{\sigma }}\right)^{2}}}$$


2.Then the tail distribution function can be wrtin in the form :

$${\displaystyle {\overline {F}}(x)\equiv \Pr[X>x]\,}=\int_{{x}}^{{\infty}} p(x) \,dx $$



For the standart normal distribution : $$\sigma =1 ,\mu=0$$
and therefore :

$${\overline {F}}(x) = \frac  {1}{{\sqrt  {2\pi }}}\int _{{x}}^{{\infty }}e^{-\frac{1}{2}x^2} dx < {\frac  {1}{{\sqrt  {2\pi }}}}\int _{{x}}^{{\infty }}e^{-\frac{1}{2}x^2}x  dx= $$

According to the Newton-Leibniz theorem

$$ =-\frac{1}{\sqrt{2\pi} }(e^{-\frac{1}{2}x^2}) \Big|_x^\infty = \frac{1}{\sqrt{2\pi} }e^{-\frac{1}{2}x^2}       $$

Therefore for all $$\alpha >0$$


$$e^{\alpha x}P[X>x]\leq e^{\alpha x}   {\frac  {1}{{\sqrt  {2\pi }}}}e^{-\frac{1}{2}x^2} = $$

$$={\frac  {1}{{\sqrt  {2\pi }}}}e^{{-\frac{1}{2}x^2}+\alpha x}{ \to 0} $$
$${x \to \infty}$$




That means thE normal distributionand has no heavy tail

The first roll, $X_0$, is $U(1...6)$, i.e. $P(X_0=i)=\pi_0(i)=1/6, \forall i\in{\{1,..,6\}}$<br>
After i rolls are determined the $i+1^{st}, X_{i+1},$ is drawn according to the row that corresponds to the value of $X_i$ in the matrix $T$ below. <br>
In other words, $T$ is the transition matrix of a Markov chain and the initial distribution is uniform.

\begin{equation*}
T = 
\begin{pmatrix}
0.4 & 0.2 & 0.1 & 0 & 0.1 & 0.2 \\
0.2 & 0.4 & 0.2 & 0.1 & 0 & 0.1 \\
0.1 & 0.2 & 0.4 & 0.2 & 0.1 & 0 \\
0 & 0.1 & 0.2 & 0.4 & 0.2 & 0.1 \\
0.1 & 0 & 0.1 & 0.2 & 0.4 & 0.2 \\
0.2 & 0.1 & 0 & 0.1 & 0.2 & 0.4
\end{pmatrix}
\end{equation*}
